```python
informacion1 = {
    'Rosaura': {'Trabajo A': 60,'Trabajo B': 70},
    'Juanita': {'Trabajo A': 90,'Trabajo B': 40},
    'Carmen': {'Trabajo A': 95,'Trabajo B': 100},
    'Pablo': {'Trabajo A': 0,'Trabajo C': 0},
    'Kelly': {'Trabajo C': 50}   
}   

informacion2 = {
    'Ana': {'Trabajo A':50,'Trabajo B':100},
    'Angela': {'Trabajo A':10,'Trabajo B':40},
    'Andres': {'Trabajo A':55,'Trabajo B':99},
    'Juan': {'Trabajo A':60,'Trabajo C': 0},
    'Paola': {'Trabajo C':10}
}

informacion3 = {
    'Pepe': {'Trabajo A':45, 'Trabajo B':90,
    'Trabajo C':100},
    'Tomas': {'Trabajo A':100,'Trabajo B':100},
    'Lucas': {'Trabajo C':10},
    'Jimena': {'Trabajo A':50,'Trabajo C': 0},
    'Paola': {'Trabajo A':10}
}

def analisisEstadistico(informacion: dict)->dict:

    estadisticas = {
        'Trabajo A': {'mejor':'', 'mayor':0, 'peor':'', 'menor':0, 'promedio':0, 'cantidad':0, 'total':0},
        'Trabajo B': {'mejor':'', 'mayor':0, 'peor':'', 'menor':0, 'promedio':0, 'cantidad':0, 'total':0},
        'Trabajo C': {'mejor':'', 'mayor':0, 'peor':'', 'menor':0, 'promedio':0, 'cantidad':0, 'total':0}
    }

    listaNotasA = []
    listaNotasB = []
    listaNotasC = []
    listaEstudiantesA = []
    listaEstudiantesB = []
    listaEstudiantesC = []
    promedioA=0
    promedioB=0
    promedioC=0
    na=0
    nb=0
    nc=0
    totalA=0
    totalB=0
    totalC=0

    for estudiantes, trabajos in informacion.items():
        if trabajos.get('Trabajo A')!=None:
            listaNotasA.append(trabajos.get('Trabajo A'))
            listaEstudiantesA.append(estudiantes)
        if trabajos.get('Trabajo B')!=None:
            listaNotasB.append(trabajos.get('Trabajo B'))
            listaEstudiantesB.append(estudiantes)
        if trabajos.get('Trabajo C')!=None:
            listaNotasC.append(trabajos.get('Trabajo C'))
            listaEstudiantesC.append(estudiantes)


    for i in listaNotasA:
        if i!=None:
            na=na+1
            promedioA=promedioA+i
            totalA=promedioA
    promedioA = round(promedioA/na, 2)

    for i in listaNotasB:
        if i!=None:
            nb=nb+1
            promedioB=promedioB+i
            totalB=promedioB
    promedioB = round(promedioB/nb, 2)


    for i in listaNotasC:
        if i!=None:
            nc=nc+1
            promedioC=promedioC+i
            totalC=promedioC
    promedioC = round(promedioC/nc, 2)




    estadisticas['Trabajo A']['mayor']=max(listaNotasA)
    estadisticas['Trabajo B']['mayor']=max(listaNotasB)
    estadisticas['Trabajo C']['mayor']=max(listaNotasC)

    estadisticas['Trabajo A']['menor']=min(listaNotasA)
    estadisticas['Trabajo B']['menor']=min(listaNotasB)
    estadisticas['Trabajo C']['menor']=min(listaNotasC)

    estadisticas['Trabajo A']['promedio']=promedioA
    estadisticas['Trabajo B']['promedio']=promedioB
    estadisticas['Trabajo C']['promedio']=promedioC

    estadisticas['Trabajo A']['cantidad']=na
    estadisticas['Trabajo B']['cantidad']=nb
    estadisticas['Trabajo C']['cantidad']=nc

    estadisticas['Trabajo A']['total']=totalA
    estadisticas['Trabajo B']['total']=totalB
    estadisticas['Trabajo C']['total']=totalC

    #Estudiantes con las mejores notas por trabajo
    estadisticas['Trabajo A']['mejor']=listaEstudiantesA[listaNotasA.index(max(listaNotasA))]
    estadisticas['Trabajo B']['mejor']=listaEstudiantesB[listaNotasB.index(max(listaNotasB))]
    estadisticas['Trabajo C']['mejor']=listaEstudiantesC[listaNotasC.index(max(listaNotasC))]

    #Estudiantes con las peores notas por trabajo
    estadisticas['Trabajo A']['peor']=listaEstudiantesA[listaNotasA.index(min(listaNotasA))]
    estadisticas['Trabajo B']['peor']=listaEstudiantesB[listaNotasB.index(min(listaNotasB))]
    estadisticas['Trabajo C']['peor']=listaEstudiantesC[listaNotasC.index(min(listaNotasC))]
  
    return estadisticas

analisisEstadistico(informacion1)
print('\n')
analisisEstadistico(informacion2)
print('\n')
analisisEstadistico(informacion3)
```
