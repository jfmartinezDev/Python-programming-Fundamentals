```python
import pandas as pd

def studentPerformance (rutaDelArchivo:str) -> dict:

    try: dataFrame = pd.read_csv(rutaDelArchivo)

    except: return "Opps algo salio mal, revise nuevamente"


    output = {

        'percentByGender': {'percentMale': 0.0, 'percentFemale':0.0 },

        'meanByCourse': {'math score': 0.0, 'writing score': 0.0, 'reading score': 0.0 },

        'meanByGender': {
            'math score': {'female': 0.0, 'male': 0.0 }, 
            'writing score': {'female': 0.0,'male': 0.0 }, 
            'reading score': {'female': 0.0, 'male': 0.0 }    
                        },

        'meanByPreparation': {
            'completed': {'math score': 0.0,'writing score': 0.0, 'reading score': 0.0 }, 
            'none': {'math score': 0.0, 'writing score': 0.0, 'reading score': 0.0 }          
                             }

    }

    gender_count = dataFrame['gender'].value_counts()

    n_female = gender_count[0]
    n_male = gender_count[1]


    output['percentByGender']['percentMale'] = round(((n_male/(n_male+n_female))*100), 2)
    output['percentByGender']['percentFemale'] = round(((n_female/(n_female+n_male))*100), 2)
    output['meanByCourse']['math score']  = round(dataFrame['math score'].mean(),2)
    output['meanByCourse']['reading score']  = round(dataFrame['reading score'].mean(),2)
    output['meanByCourse']['writing score']  = round(dataFrame['writing score'].mean(),2)




    gender_score = dataFrame[['gender','math score', 'reading score', 'writing score']].groupby(['gender']).agg(['mean']).round(2)


    output['meanByGender']['math score']['female'] = round(gender_score['math score']['mean'][0],2)
    output['meanByGender']['math score']['male'] = round(gender_score['math score']['mean'][1],2)

    output['meanByGender']['writing score']['female'] = round(gender_score['writing score']['mean'][0],2)
    output['meanByGender']['writing score']['male'] = round(gender_score['writing score']['mean'][1],2)

    output['meanByGender']['reading score']['female'] = round(gender_score['reading score']['mean'][0],2)
    output['meanByGender']['reading score']['male'] = round(gender_score['reading score']['mean'][1],2)




    test_preparation_score = dataFrame[['test preparation course','math score', 'reading score', 'writing score']].groupby(['test preparation course']).agg(['mean']).round(2)


    output['meanByPreparation']['completed']['math score'] = round(test_preparation_score['math score']['mean'][0],2)
    output['meanByPreparation']['completed']['writing score'] = round(test_preparation_score['writing score']['mean'][0],2)
    output['meanByPreparation']['completed']['reading score'] = round(test_preparation_score['reading score']['mean'][0],2)


    output['meanByPreparation']['none']['math score'] = round(test_preparation_score['math score']['mean'][1],2)
    output['meanByPreparation']['none']['writing score'] = round(test_preparation_score['writing score']['mean'][1],2)
    output['meanByPreparation']['none']['reading score'] = round(test_preparation_score['reading score']['mean'][1],2)

    return output
```
